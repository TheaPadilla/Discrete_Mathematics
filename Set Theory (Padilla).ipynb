# 1.) Proving Identities

 1. $A \setminus B = A \cap B^C$
<br>To show that $A \setminus B = A \cap B^C$, we must consider the ff.
- $A \setminus B$ comprises elements in $A$ that are absent in $B$.
- $A \cap B^C$ contains elements in $A$ that are also in the complement of $B$.
<br>$Proof:$
<br>Any $x$ in ${A}\setminus {B}$ implies $x$ is in $A$ but not in $B$.
<br>Any $x$ in ${A} \cap B^C$ implies $x$ is in $A$ and in the complement of $B$.
<br>**Since these conditions are equivalent, we can conclude that ${A}\setminus{B}={A}\cap B^C$**
2. $B = (B \cap A) \cup (B \cap A^C)$
<br>$Proof:$

From the complement law, we know that if you take the union of a set $A$ and its complement ($A^C$), you get the universal set ($\mathcal{U}$).This can be expressed as $A\cup A^C =\mathcal{U}$
